```python
import pandas as pd
```

```python
order_df = pd.read_csv(r"/Users/saideexith/datacleaning/Amazon Dashboard/order_list.csv")
```

```python
price_df = pd.read_csv(r"/Users/saideexith/datacleaning/Amazon Dashboard/price_list.csv")
```

```python
order_df
```

```python
merged_df = order_df.merge(price_df , how = "left" , on = 'ASIN')
```

```python
merged_df
```

```python
merged_df['Total_cost'] = merged_df['sales_price']*merged_df['Quantity']
```

```python
merged_df
```

```python
merged_df[['Year', 'Month', 'Day']] = merged_df['Sales Data'].str.split('-', expand=True)
```

```python
merged_df['Month'] = merged_df['Month'].astype(int)
```

```python
merged_df
```

```python
merged_df.drop(columns = 'Year')
```

```python
merged_df['Day'] = merged_df['Date'].dt.day
```

```python
merged_df
```

```python
merged_df['Sales Data'] = merged_df['Date']
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns = ['Date' , 'Year'])
```

```python
merged_df
```

```python
merged_df.sort_values(by='Sales Data')
```

```python
df_march = merged_df[merged_df['Month'] == 3]
```

```python
df_march
```

```python
expenseMarch = df_march['Total_cost'].sum()
```

```python
expenseMarch
```

```python
cancelled_march = (df_march['Quantity'] == 0).sum()
```

```python
cancelled_march
```

```python
grouped_marchdf = df_march.groupby('ship-state')['Total_cost'].sum().reset_index()
```

```python
grouped_marchdf
```

```python
grouped_marchdf = df_march.groupby('ship-state')[['Total_cost' , 'Quantity']].sum().reset_index()
```

```python
grouped_marchdf
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for Orders
ax1.bar(grouped_marchdf['ship-state'], grouped_marchdf['Quantity'], color='b', alpha=0.6, label='Orders')
ax1.set_xlabel('State')
ax1.set_ylabel('Orders', color='b')
ax1.tick_params(axis='y', labelcolor='b')

plt.xticks(rotation=90)

ax2 = ax1.twinx()
ax2.bar(grouped_marchdf['ship-state'], grouped_marchdf['Total_cost'], color='g', alpha=0.6, label='Cost')
ax2.set_ylabel('Cost', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Title and Legend
plt.title('State vs. Cost and Orders')
fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
```
